这个代码转换sample rate

```python
import torchaudio
import os

# 原始音频文件夹路径
input_folder = "/content/denoiser_P/noisy48"
# 重新采样后的音频文件夹路径
output_folder = "/content/denoiser_P/noisy"

# 确保输出文件夹存在
os.makedirs(output_folder, exist_ok=True)

# 列出原始音频文件夹中的所有文件
audio_files = os.listdir(input_folder)

# 遍历每个音频文件并重新采样
for audio_file in audio_files:
    if audio_file.endswith(".wav"):
        # 构建输入和输出文件的完整路径
        input_file = os.path.join(input_folder, audio_file)
        output_file = os.path.join(output_folder, audio_file)

        # 加载音频文件
        waveform, sample_rate = torchaudio.load(input_file)

        # 将音频重新采样为16000 Hz
        resampled_waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)(waveform)

        # 保存重新采样后的音频文件
        torchaudio.save(output_file, resampled_waveform, 16000)
        print(f"Resampled and saved: {output_file}")


# 原始音频文件夹路径
input_folder = "/content/denoiser_P/clean48"
# 重新采样后的音频文件夹路径
output_folder = "/content/denoiser_P/clean"

# 确保输出文件夹存在
os.makedirs(output_folder, exist_ok=True)

# 列出原始音频文件夹中的所有文件
audio_files = os.listdir(input_folder)

# 遍历每个音频文件并重新采样
for audio_file in audio_files:
    if audio_file.endswith(".wav"):
        # 构建输入和输出文件的完整路径
        input_file = os.path.join(input_folder, audio_file)
        output_file = os.path.join(output_folder, audio_file)

        # 加载音频文件
        waveform, sample_rate = torchaudio.load(input_file)

        # 将音频重新采样为16000 Hz
        resampled_waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)(waveform)

        # 保存重新采样后的音频文件
        torchaudio.save(output_file, resampled_waveform, 16000)
        print(f"Resampled and saved: {output_file}")
```
